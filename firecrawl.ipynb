```python
from firecrawl import Firecrawl

firecrawl = Firecrawl(api_key="fc-")

URL = "https://www.musinsa.com/content/1448323794271830123?utm_source=/main/sneaker/recommend&utm_medium=main_display_big_banner&utm_campaign=1"

# Scrape a website:
doc = firecrawl.scrape(URL, formats=["markdown", "html"])
print(doc)
```

```python
from pprint import pprint

with open('./my-crawl.md', 'w', encoding='utf-8') as f:
    f.write(doc.markdown)

pprint(doc.markdown)
```
