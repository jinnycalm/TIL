# 05_함수
```
def 함수_이름(매개변수-input):

    [함수로 묶을 코드]

    return 반환값-output
```

````python
# 함수 정의(define)
def plus_one(num):
    return num + 1

# 함수 호출(call)
plus_one(1)
````

````python
# n번 인사하는 함수
# 정의
def hello_ntimes(n):
    # for-매개변수에 들어간 값을 사용하지 않을 때 언더바를 사용.
    for _ in range(n):
        print('안녕하세요.')

# 호출
hello_ntimes(5)


# 정의할때 사용하는 n은 매개변수(parameter), 5, 10 등은 인자(argument)
````

````python
# 3제곱 해주는 함수
# name: cube, in: 숫자1개, out: 숫자의 세제곱

def cube(num):
    return num **3

cube(3)
````

````python
# 원기둥의 부피를 구하는 함수
def cylinder(num1, num2):
    pi = 3.14
    return 3.14 * (num1 **2) * num2

cylinder(3, 4)
round(cylinder(3, 4))
````

````python
# 리스트 내부 숫자들의 평균을 구해주는 함수
def get_avg(new_list):
    
    sum = 0
    for a in new_list:
        sum += a
    avg = sum / len(new_list)
    return avg


get_avg([1, 2, 3, 4, 5])

````

````python
def check_pw(pwpw):
    result = False
    if len(pwpw) >= 8:
        for char in pwpw:
            if char in "0123456789":
                result = True
    return result


check_pw('sd2323kd23')
````
