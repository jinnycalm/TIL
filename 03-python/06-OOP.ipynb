```python
# phthon에서 클래스 이름은 PascalCase로 짓는다.

class Person:
    # 1. 특성(attribute)
    height = 165
    weigh = 55
    age = 20
    iq = 100
    name = '인간'

    # 2. 행동(method)
    def greeting(self):
        return f'안녕하세요, {self.name}입니다. {self.age}살 입니다.'

    def study(self):
        self.iq += 5
```

```python
# Person 개념적 대상을 -> 실제 생성
p1 = Person()
p2 = Person()
```

```python
# 특성
p1.age = 30
p1.name = '김로봇'
p1.weight = 80
p1.height = 230
# 행동
p1.greeting()
```

```python
# 특성
p2.age = 23
p2.name = '박사람'
p2.weight = 45
p2.height = 155
# 행동
p2.greeting()
```

```python
# 개념적인 인간을 뽑아내고 p1, p2를 각각 속성을 바꿔서 세팅하였음. 완전 다른 존재가 된거임.
# 개념이 class, 실존하는 실체가 instance라는 이름으로 불리고 있는 것.

# OOP 프로그래밍 탄생 이유 => 현실세계 반영하기 위해서.
# 실존하는 인물들은 상위에 idea라는 것이 있음.

# 순수 객체 지향
# 모든 값은 다 객체다. 
```
