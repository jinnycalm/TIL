```python
# 조건 - if-else

n = 10

if n > 0:
    print('0보다 크다')
else:
    print('0보다 작거나 같다')
```

```python
# 조건 - if-elif-else
n = -10

if n > 0:
    print('양수')
elif n == 0:
    print('0입니다.')
elif n < 0:
    print('음수')
else:
    print('숫자가 아닙니다.')
```

```python
score = 400

if score >= 1000:
    print('올바른 점수가 아닙니다.')
elif score >= 900:
    print('LV.4')
elif score >= 750:
    print('LV.3')
elif score >= 600:
    print('LV.2')
elif score >= 400:
    print('LV.1')
else:
    print('불합격')
```

```python
# 우수학생판별
status = {
    'score': 60,
    'absent': 5
}
# 1. 점수가 70을 넘는 학생이면 우수학생
# 2. 우수학생 중에 결석일이 5일 미만이면 최우수학생
# 3. (우수학생아닌 학생중) 결석일이 5일 미만이면 성실학생
# 4. 나머지는 일반학생

if status['score'] > 70:
    if status['absent'] < 5:
        print('최우수학생')
    else:
        print('우수학생')
elif status['absent'] < 5:
    print('성실학생')
else:
    print('일반학생')
```
