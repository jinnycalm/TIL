```python
# 문제 1. 점수 평균과 합격 판별
# 학생 점수 리스트가 있다.
# scores = [80, 75, 90, 60, 100]평균을 구하시오.평균이 80 이상이면 "합격" 아니면 "불합격" 출력
```

```python
scores = [80, 75, 90, 60, 100]

total = 0
i = 0
while i < 5:
    total += scores[i]
    i += 1
score_avg = total/len(scores)
if score_avg >= 80:
    print('합격')
else:
    print('불합격')
```

```python

#  문제 2. 최대값 찾기
# 숫자 리스트가 있다.
# numbers = [3, 7, 2, 9, 5]가장 큰 값을 찾아 출력하시오.
# (힌트: max() 없이 직접 비교해보기)
```

```python
numbers = [3, 7, 2, 9, 5]

# 우선 최댓값 후보로 첫번째 숫자를 잡는다.
maximum = numbers[0]

for num in numbers:
    if num > maximum:
        maximum = num

print(maximum)

```

```python
#  문제 4. 이름 존재 여부 확인
# target = "민수"
# names = ["철수", "영희", "민수", "지영"]변수 target 에 저장된 이름이 리스트 안에 있으면
#  "존재합니다" 아니면 "없습니다" 출력
```

```python
target = "민수"
names = ["철수", "영희", "민수", "지영"]

print(target in names)

result = '없습니다'
for name in names:
    if name == target:
        result = '있습니다.'

print(result)

```

```python
#  문제 5. 딕셔너리 값 접근
# student = {
#     "name": "철수",
#     "age": 20,
#     "score": 85
# }점수가 80 이상이면 "우수 학생" 출력아니면 "보통 학생" 출력
```

```python
#  문제 6. 재고 확인 시스템

# inventory = {
#     "apple": 10,
#     "banana": 0,
#     "orange": 5
# }각 과일의 재고가
# 0이면 → "품절"
# 1 이상이면 → "판매 가능"
# 출력하시오.
```

```python
inventory = {
    "apple": 10,
    "banana": 0,
    "orange": 5
}


for fruit in inventory:
    if inventory[fruit] > 0:
        available = f'{fruit} 판매가능'
        print(available)
    else:
        not_available = f'{fruit} 품절'
        print(not_available)


```

```python
#  문제 7. 리스트 안 딕셔너리 평균 구하기
# students = [
#     {"name": "철수", "score": 80},
#     {"name": "영희", "score": 90},
#     {"name": "민수", "score": 70}
# ]전체 평균 점수 구하기평균 이상인 학생 이름 출력
```

```python
students = [
    {"name": "철수", "score": 80},
    {"name": "영희", "score": 90},
    {"name": "민수", "score": 70}
]

sum = 0

for nums in students:
    sum += nums["score"]
    avg = sum/len(students)
    
    if nums["score"] >= avg:
        print(nums['name'])
    
```

```python
#  문제 8. 등급 매기기
# 점수 리스트:
# scores = [95, 82, 67, 74, 88]각 점수에 대해
# 90 이상 → A80 이상 → B70 이상 → C그 외 → D등급 리스트를 새로 만들어 출력하시오.
```

```python
scores = [95, 82, 67, 74, 88]
new_list = []

i = 0
for i in range(5):
    if scores[i] >= 90:
        new_list.append('A')
    elif scores[i] >80:
        new_list.append('B')
    elif scores[i] >70:
        new_list.append('C')
    else:
        new_list.append('D')

print(new_list)
```

```python
#  문제 9. 장바구니 총 금액 계산
# cart = [
#     {"item": "노트북", "price": 1000000, "qty": 1},
#     {"item": "마우스", "price": 20000, "qty": 2}
# ]총 결제 금액 계산100만원 이상이면 "고액 구매자" 출력
```

```python
cart = [
    {"item": "노트북", "price": 1000000, "qty": 1},
    {"item": "마우스", "price": 20000, "qty": 2}
]

# print(cart[0]['price'])

sum = 0

for num in cart:
    # print(cart[num]['price'])
    item_price = num['price']
    item_qty = num['qty']
    sub_amount = item_price * item_qty
    sum += sub_amount

if sum >= 1000000:
    print('고액 구매자')
else:
    print('일반 구매자')
```

```python

#  문제 10. 비밀번호 검사기
# password = "abc123"조건:
# 길이가 8자 이상이고
# 숫자가 포함되어 있으면 "사용 가능"
# 아니면 "사용 불가"
# 출력하시오.
# (힌트: len(), in  사용)
```

```python
password = "abd223233s"

result = '사용불가'

if len(password) >= 8:
    for pw in password:
        if pw in "0123456789":
            result = '통과'

print(result)
```
